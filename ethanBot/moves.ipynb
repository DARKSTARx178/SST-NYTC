```python
from ugot import ugot
got = ugot.UGOT()
got.initialize('192.168.88.1')
```

crashout


```python
import time

angle = 0
claw_open = True

try:
    while True:
        got.mechanical_joint_control(0, angle, 90, 10)
        angle = (angle + 15) % 180

        # Claw open/close toggle
        if claw_open:
            got.mechanical_clamp_close()
        else:
            got.mechanical_clamp_release()  

        claw_open = not claw_open

except KeyboardInterrupt:
    got.mecanum_move(0, 0, 0)
```

dance

```python
import time

angle = 0
claw_open = True

# Optional: define a “beat” pattern (frequency in Hz, duration in ms)
beats = [
    (800, 100),  # beep 1
    (1000, 100), # beep 2
    (1200, 100), # beep 3
    (1000, 100), # beep 4
]
beat_index = 0

try:
    while True:
        # 🔄 Move in a circle
        got.mecanum_translate_speed(angle=0, speed=80)
        got.mecanum_turn_speed(turn=3, speed=80)

        # 🤖 Arm movement
        got.mechanical_single_joint_control(joint=1, angle=angle, duration=100)
        got.mechanical_single_joint_control(joint=2, angle=angle % 90, duration=100)
        got.mechanical_single_joint_control(joint=3, angle=angle % 90, duration=100)
        angle = (angle + 20) % 270

        # 🦾 Claw toggle
        if claw_open:
            got.mechanical_clamp_release()
        else:
            got.mechanical_clamp_close()
        claw_open = not claw_open

        # 🔊 Play a beat (loop through beats)
        freq, dur = beats[beat_index]
        got.play_tone(freq, dur)
        beat_index = (beat_index + 1) % len(beats)

        time.sleep(0.2)  # sync speed to music

except KeyboardInterrupt:
    # Reset robot
    got.mechanical_joint_control(0, 90, 90, 10)    
    got.mechanical_clamp_close()
    got.mecanum_stop()                  
```
